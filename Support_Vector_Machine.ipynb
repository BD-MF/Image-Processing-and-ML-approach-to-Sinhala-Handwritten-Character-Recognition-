```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import pickle
from ast import literal_eval
%matplotlib inline
```

```python
data = pd.read_csv("C:/CHARACTOR_RECOGNITION/NEW_DATA_FINAL.csv")
data
```

```python
ex =data.drop('Group', axis=1)
ex =ex.drop('label', axis=1)
ex =ex.drop('unicode', axis=1)
X =ex.drop('Num', axis=1)
y = data['unicode']
X
```

```python
X.dropna()
```

```python
df = pd.DataFrame(X)
```

```python
df['LR'] = df['LR'].astype(str).apply(literal_eval)
df['RL'] = df['RL'].astype(str).apply(literal_eval)
df['TB'] = df['TB'].astype(str).apply(literal_eval)
df['BT'] = df['BT'].astype(str).apply(literal_eval)
df['Density'] = df['Density'].astype(str).apply(literal_eval) 
```

```python
dis = []
length = len(df)
#print(length)
for i in range(0,length):
    #print(i)
    arr = np.array(df['LR'][i]+df['RL'][i]+df['TB'] [i]+df['BT'][i] )
    arr= np.append(arr,df['Density'][i])
    arr= np.append(arr,df['Ratio0'][i])
    arr= np.append(arr,df['Ratio1'][i])
    arr= np.append(arr,df['Ratio2'][i])
    arr= np.append(arr,df['Ratio3'][i])
    arr= np.append(arr,df['Ratio4'][i])
    arr= np.append(arr,df['Ratio5'][i])
    arr= np.append(arr,df['Ratio6'][i])
    arr= np.append(arr,df['Ratio7'][i])
    dis.append(arr)
    
#dis
```

```python
X_train, X_test, y_train, y_test = train_test_split(dis, y, test_size = 0.20)
#X_train
```

```python
#y_test
```

```python
mm_scaler = preprocessing.MinMaxScaler()
X_train_minmax = mm_scaler.fit_transform(X_train)
mm_scaler.transform(X_test)
```

```python
#X_train
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear', C=1.0, probability=False, degree=3, coef0=0.0, 
cache_size=100.0, gamma=0.00666666666667)
model = svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# get the accuracy
accuracy_score(y_test, y_pred)*100
```
